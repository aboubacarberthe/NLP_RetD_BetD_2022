1. TF-IDF

Une méthode statistique qui évalue la pertinence d'un mot dans un document. Ceci en multipliyant deux matrices :

* Combien de fois le mot apparait dans un document

* L'inverse du nombre de document dans lesquels le mot est présent

En résumé : TF-IDF définit l'importance d'un mot en prenant en compte : 
* l'importance de ce mot dans un seul document 

* L'importance du mot dans tous les documents

## Association 

==> Chercher les mots métiers qui apparaissent le plus souvent ensemble. Combien de fois chaque mot métier est présent sur chaque ticket, l'ensmble des tickets ? 

```python
### Contexte functions 
import numpy as np
import re

def get_context(text, word_list,context):
    """
    Get n words before and after a given word in sentence

            Parameters:
                    text (column)  : the text that contains our word
                    word_list(list) : list of word that we would like to find context
                    context (int): the number of words before and after the given word from word_list

            Returns: 

                    a list with all outputs
    """   
    words = re.findall(r'\w+', text)
    tab = []
    w= []
    for t in word_list:
        matches = (i for (i,w) in enumerate(words) if w.lower() == t)
        for index in matches:
            if index < context //2:
                tab.append(words[0:context+1])
            elif index > len(words) - context//2 - 1:
                tab.append(words[-(context+1):])
            else:
                tab.append(words[index - context//2:index + context//2 + 1])
    return tab
def get_context_of_word(texte,word):
    tab = []
    for i in range(len(texte)):
        tab.append(get_context(texte[i] , [word],10))
    l = [item for sublist in tab for item in sublist]
    return [" ".join(i for i in l[k]) for k in range(len(l))]
    
```

```python
def get_ngram(texte,word, n):
    """
    Get n words before and after a given word in sentence

            Parameters:
                    text (column)  : the text that contains our word
                    word_list(list) : list of word that we would like to find context
                    context (int): the number of words before and after the given word from word_list

            Returns: 

                    a list with all outputs
    """   
    
    n_grams = [get_context(texte[i], list(mots_metiers["words"]),n) for i in range(len(texte))]
    n_ = [item for sublist in n_grams for item in sublist]
    return [i for i in n_ if word in i]

```
